```python
print("Total Sales:", df[' Sales'].sum())
print("Total Profit:", df['Profit'].sum())
```

```python
profit_margin = (df['Profit'].sum() / df[' Sales'].sum()) * 100

print("Profit Margin:", round(profit_margin, 2), "%")
```

```python
df.groupby('Country')[' Sales'].sum().sort_values(ascending=False)
```

```python
df.groupby('Product')['Profit'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

top_country = df.groupby('Country')[' Sales'].sum().sort_values(ascending=False)

top_country.plot(kind='bar')
plt.title('Sales by Country')
plt.ylabel('Sales')
plt.show()
```

```python
top_product = df.groupby('Product')['Profit'].sum().sort_values(ascending=False)

top_product.plot(kind='bar')
plt.title('Profit by Product')
plt.ylabel('Profit')
plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)

segment_sales.plot(kind='bar')
plt.title('Sales by Segment')
plt.ylabel('Sales')
plt.show()
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)

segment_sales.plot(kind='bar')
plt.title('Sales by Segment')
plt.ylabel('Sales')
plt.show()
```

```python
country_profit = df.groupby('Country')['Profit'].sum().sort_values(ascending=False)

country_profit.plot(kind='bar')
plt.title('Profit by Country')
plt.ylabel('Profit')
plt.show()
```

```python
monthly_sales = df.groupby('Month Name')['Sales'].sum()

monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales')
plt.show()
```

```python
monthly_profit = df.groupby('Month Name')['Profit'].sum()

monthly_profit.plot(kind='line', marker='o')
plt.title('Monthly Profit Trend')
plt.ylabel('Profit')
plt.show()
```
